```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {'a': [2,3,2,7,4],
     'b': [2,1,3,4,3],
     'c': [1,1,2,3,5],
     }
)
```

```python
df.sort_index(ascending=False)
```

값을 기준으로 정렬

```python
df.sort_values( by = ['a'], inplace=True, ascending=False) # ascending 내림차순
df
```

- sort_index => index 기준 정렬
  - by => 기준이 되는 컬럼이름
  - ascending =>True : 오름차순 / False : 내림차순
  - inplace => True: 현재 수행결과를 저장할 때 / False: 임시
- sort_value => 특정 칼럼을 기준으로 정렬
  - by => 기준이 되는 컬럼이름
  - ascending =>True : 오름차순 / False : 내림차순
  - inplace => True: 현재 수행결과를 저장할 때 / False: 임시

# 결측지 처리

```python
# None, NAN, NaN
# np.nan : 임의의 결측값
import numpy as np
df = pd.DataFrame(
    {'a':[1,1,3,4,5],
     'b':[2,3,np.nan,3,4],
     'c':[3,4,7,6,4]}
)
df
```

- 결측치 유무

```python
df.isnull() 
# 결측치 확인
```

```python
# Nan의 갯수 확인
df.isnull().sum()
```

```python
# 결측치를 포함한 행 삭제
df.dropna()
# 저장x

# df.dropna(inplace=True) 저장o

# 결측치를 포함한 컬럼 삭제 axis = 0 가로 / axis = 1 세로
# df.dropna(axis=1)
```

```python
df
# 결측치값 대체
# df.fillna(대체하고자 하는 값)

df.fillna(10) #Nan을 10으로 변경
```

```python
# 결측치값 이전 혹은 값으로 대체
df2 = pd.DataFrame(
    {'a':[1,1,3,4,np.nan],
     'b': [2,3,np.nan,np.nan,4],
     'c': [np.nan,4,1,1,4]
    }
)
```

```python
# 특정 위치에 임의의 nan 값 지정할 때 사용
df2['a'].loc[0]=np.nan
df2
```

```python
# 결측치 체울시 채우는 갯수 선택적으로 제한하는 경우
df2.fillna(method='ffill', limit=1)
```

```python
df2.fillna(method='ffill')
```

```python
# bfill ffill로 결측치 제거 수행

df2.fillna(method='bfill')
```

```python
df2.fillna(method='ffill', inplace=True)
df2.fillna(method='bfill', inplace=True)
df2
```

# apply, map을 화용한 데이터 변화

```python
df=pd.DataFrame({
    'a':[1,2,3,4,5]
}
)
df
```

```python
# 컬럼 b를 생성 / 조건 a가 2보다 작으면 '2미만', 4보다 작으면 '4미만', 4보다 크면 '4이상'이라는 값을 입력
```

```python
df['b']=0
df.head()
```

```python
a = df[df['a']<2]
a
```

```python
df['b'][a.index] = '2미만'
df
```

```python
# df['b'] = (df[df['a']>=2]) & (df[df['a']<4])
df['b'][a.index] = '4미만'
df
```

```python
def cls_function(x):
  if x<2:
    return '2미만'
  elif x<4:
    return '4미만'
  else:
    return '4이상'
```

```python
df['c'] = df['a'].apply(cls_function)
df
```

```python
y=2
z=4
df['d'] = df['a'].apply(lambda x : '2미만' if (x<2) else ('4미만' if (x<4) else '4이상'))
df
```

- 데이터 병합

- 1) 상하 결합

```python
df1 = pd.DataFrame({'a':[1,2,3], 'b':[11,12,13], 'c':[21,22,23]})
df2 = pd.DataFrame({'a':[4,5,6], 'b':[13,15,16], 'c':[24,25,26]})
```

```python
# 결합시 사용하는 함수는 concat() 사용
```

```python
pd.concat([df1,df2])
```

```python
# index 초기화
pd.concat([df1, df2], ignore_index=True)
```

```python
df1 = pd.DataFrame({'a':[1,2,3], 'b':[11,12,13], 'd':[21,22,23]})
df2 = pd.DataFrame({'a':[4,5,6], 'b':[13,15,16], 'c':[24,25,26]})
pd.concat([df1,df2])
# 상하 결합
```

```python
# innver join
df1 = pd.DataFrame({'a':[1,2,3], 'b':[11,12,13], 'd':[21,22,23]})
df2 = pd.DataFrame({'a':[4,5,6], 'b':[13,15,16], 'c':[24,25,26]})
pd.concat([df1,df2], join='inner')
# 같은 컬럼만
```

- 2)좌우 결합

```python
pd.concat([df1, df2], axis = 1)
# columns 가로로 병합
```

```python
# merge 함수
# 특정 기준이나 결합 방식을 정의랑 때 사용
df1 = pd.DataFrame({'id':[1,2,3], '성별':['F','M','F'], '나이':[20,30,40]})
df2 = pd.DataFrame({"id":[1,2,3], '키':[160.5,170.3,180.1], "몸무게":[45.1,50.3,72.1]})
```

```python
df1
```

```python
df2
```

```python
# 아이디를 기준으로 합칠 때 merge

pd.merge(df1, df2, how='left', on='id')
```

```python
df1 = pd.DataFrame({'id':[1,2,3,4,5], '성별':['F','M','F','M','F'], '나이':[220,30,40,25,42]})
df2 = pd.DataFrame({'id':[3,4,5,6,7], '키':[160.5,170.3,180.1,142.3,156.7], '몸무게':[45.1,50.3,72.1,38,42]})
```

```python
pd.merge(df1,df2, how='left', on='id')
```

- 키, 몸무게, 성별, 나이 정보가 모두 확인된 유저들의 정보를 출력하시오

```python
dz = pd.merge(df1,df2, how='left', on='id').dropna()
# 결측치 제거
dz
```

```python
# 전체 데이터를 결합해서 출력할 때
pd.merge(df1,df2, how='outer', on='id')
```
